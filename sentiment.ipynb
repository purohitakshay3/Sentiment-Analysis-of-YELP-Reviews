```python
import os
import glob
from textblob import TextBlob

nlst = open('negative-words.txt', 'r').read().split(';')
neg_list = lst[-1].split()

plst = open('positive-words.txt', 'r').read().split(';')
pos_list = lst[-1].split()

path = r'C:\Users\puroh\OneDrive\Documents\Python Scripts\Question 3\yelp100Reviews\yelp100Reviews'

affin = open('AFINN-111.txt', 'r')
affin_list = affin_content = affin.read().split('\n')
my_dict = {}
final_dict = {}
for i in affin_list:
    temp_list = i.split('\t')
    my_dict = {temp_list[0]:int(temp_list[1])}
    if final_dict == {}:
        final_dict = my_dict
    else:
        final_dict.update(my_dict)

print("Review\tAFFIN-Sentiment\tNet Sentiment Score ")
    
for filename in glob.glob(os.path.join(path, '*.txt')):
    basename_list = os.path.basename(filename).split('.')
    #print basename_list
    data = open(filename)
    content = ''
    content = data.read()
    content_list = []
    content_list = content.split()
    pos_sent = 0
    neg_sent = 0
    sent = 0
    net_sent = 0
    for word in content_list:
        
        if word in final_dict:           #For cumulative sentiments
            sent += final_dict[word]
            
        if word in pos_list:             #For net score
            pos_sent += 1
        elif word in neg_list:
            neg_sent += 1
    
    net_sent = pos_sent - neg_sent
            
    print("{}\t{}\t\t{}".format(basename_list[0], sent, net_sent))   
```
